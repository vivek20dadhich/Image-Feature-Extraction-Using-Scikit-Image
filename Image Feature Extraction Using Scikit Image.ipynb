# Converting to gray image

```python
from skimage.color import rgb2gray # converting to grayscale so that our image has only one color channel

grayimg = rgb2gray(img)
imshow(grayimg)
```

```python
grayimg.shape  # shape of grayscale image
```

```python
grayimg.size   # size of grayscale image 
               # The number of pixels in an image is the same as the size of the image for grayscale image
```

#  Pixel Features

we can find the pixel features by reshaping the shape of the image and returning the array form of the image

```python
pixel_feat1 = np.reshape(grayimg, (1000 * 750))                            # pixel feature for the gray image
pixel_feat1
```

```python
pixel_feat2 = np.reshape(img, (1000 * 750 * 3))                            # pixel feature for the colored image
pixel_feat2
```

# Edge Features

Here I used 

Sobel Kernel:
It works on creating images with emphasis on edges

Canny algorithm:
Which is an edge detection technique that uses a multi-stage algorithm to detect a wide range of edges in images

```python
from skimage import filters
from skimage import feature

# Sobel Kernel
sob = filters.sobel(grayimg)

imshow(sob, cmap='gray');
```

```python
#canny algorithm
can = feature.canny(grayimg)

imshow(can, cmap='gray');
```

# Region-Based Segmentation

```python
from skimage.exposure import histogram
hist, hist_centers = histogram(grayimg)

#Plotting the Image and the Histogram of gray values

fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].imshow(grayimg, cmap=plt.cm.gray)
axes[0].axis('off')
axes[1].plot(hist_centers, hist, lw=2)
axes[1].set_title('histogram of gray values')
```
